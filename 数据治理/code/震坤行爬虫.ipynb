```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.service import Service
from parsel import Selector
import pandas as pd
import time
```

```python
# 信息存储
data_dict = {
    "商品信息": [],
    "链接": [],
    "订货编码": [],
    "品牌": [],
    "制造商型号": [],
    "产品规格": [],
    "发货日": [],
    "含税单价": []
}
```

```python
# 模拟人工操作
options = webdriver.ChromeOptions()
service = Service(executable_path=r'D:\Python\chromedriver-win64\chromedriver.exe') 
# options.add_argument('--headless')  # 无头模式（可选）
driver = webdriver.Chrome(options=options, service=service)
driver.get("https://www.zkh.com/search.html?keywords=%E8%9E%BA%E4%B8%9D%E9%92%89&hasLinkWord=1")
# # 等待新内容加载
WebDriverWait(driver, 3)
driver.execute_script("window.scrollTo(0,10000)")
driver.execute_script("window.scrollTo(10000,0)")
driver.execute_script("window.scrollTo(0,10000)")
driver.execute_script("window.scrollTo(10000,0)")
driver.execute_script("window.scrollTo(0,10000)")
time.sleep(10)
# 找到列表元素模拟点击
element = driver.find_element(
    By.CSS_SELECTOR, 
    ".filter-list-type li.active"  # 注意两个 class 之间用`.`连接，无需空格
)
element.click()
# 滑到最底部
driver.execute_script("window.scrollTo(0,10000)")
driver.execute_script("window.scrollTo(10000,0)")
driver.execute_script("window.scrollTo(0,10000)")
driver.execute_script("window.scrollTo(10000,0)")
driver.execute_script("window.scrollTo(0,10000)")
time.sleep(3)
sel = Selector(driver.page_source)
for sel_ in sel.css(".goods-list.clearfix"):
    # 链接
    link = 'https://www.zkh.com' + sel_.css(".goods-img.tw1 a::attr('href')").get()
    data_dict["链接"].append(link)
    # 商品信息
    title = "".join(sel_.css(".tw2.goods-info .name.clamp3 ::text").getall())
    data_dict["商品信息"].append(title)
    # 订货编码
    sku = sel_.css(".tw2.goods-info .clearfix.sku-number ::text").get()
    data_dict["订货编码"].append(sku)
    # 品牌
    brand = sel_.css(".tw3.brand-name::text").get()
    data_dict["品牌"].append(brand)
    # 制造商型号
    mat = sel_.css(".tw4.material-number::text").get()
    data_dict["制造商型号"].append(mat)
    # 产品规格
    feat = sel_.css(".tw5.feature ::text").getall()
    data_dict["产品规格"].append(feat)
    # 发货日
    tm = sel_.css(".tw7.deliver-leadTime ::text").get()
    data_dict["发货日"].append(tm)
    # 含税单价
    price = "".join(sel_.css(".sku-price-wrap-new ::text").getall())
    data_dict["含税单价"].append(price)
for i in range(9):
    print(f"***************第{i+1}页开始抓取***************")
    wait = WebDriverWait(driver, 3)
    element = driver.find_element(
        By.CSS_SELECTOR, 
        "button.nextbtn"  # 注意两个 class 之间用`.`连接，无需空格
    )
    element.click()
    # 滑到最底部
    driver.execute_script("window.scrollTo(0,8000)")
    driver.execute_script("window.scrollTo(8000,0)")
    driver.execute_script("window.scrollTo(0,8000)")
    driver.execute_script("window.scrollTo(8000,0)")
    driver.execute_script("window.scrollTo(0,8000)")
    driver.execute_script("window.scrollTo(8000,0)")
    driver.execute_script("window.scrollTo(0,8000)")
    driver.execute_script("window.scrollTo(8000,0)")
    driver.execute_script("window.scrollTo(0,8000)")
    time.sleep(3)
    sel = Selector(driver.page_source)
    for sel_ in sel.css(".goods-list.clearfix"):
        # 链接
        link = 'https://www.zkh.com' + sel_.css(".goods-img.tw1 a::attr('href')").get()
        data_dict["链接"].append(link)
        # 商品信息
        title = "".join(sel_.css(".tw2.goods-info .name.clamp3 ::text").getall())
        data_dict["商品信息"].append(title)
        # 订货编码
        sku = sel_.css(".tw2.goods-info .clearfix.sku-number ::text").get()
        data_dict["订货编码"].append(sku)
        # 品牌
        brand = sel_.css(".tw3.brand-name::text").get()
        data_dict["品牌"].append(brand)
        # 制造商型号
        mat = sel_.css(".tw4.material-number::text").get()
        data_dict["制造商型号"].append(mat)
        # 产品规格
        feat = sel_.css(".tw5.feature ::text").getall()
        data_dict["产品规格"].append(feat)
        # 发货日
        tm = sel_.css(".tw7.deliver-leadTime ::text").get()
        data_dict["发货日"].append(tm)
        # 含税单价
        price = "".join(sel_.css(".sku-price-wrap-new ::text").getall())
        data_dict["含税单价"].append(price)
    print(f"---------第{i+1}页抓取成功---------------")
```

```python
df = pd.DataFrame(data_dict)
df = df.drop_duplicates(["商品信息", "链接", "订货编码", "品牌", "制造商型号"])
df
```
